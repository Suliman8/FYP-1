```python
# Categorical Feature Encoding for UNSW-NB15 Dataset
# -------------------------------------------------

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
import io
from google.colab import files
from IPython.display import display, HTML

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# Upload cleaned dataset (or use the original if you haven't completed the missing values step)
print("Please upload your UNSW-NB15 training dataset CSV file (cleaned or original)...")
uploaded = files.upload()

# Get the name of the uploaded file
filename = list(uploaded.keys())[0]
print(f"Processing file: {filename}")

# Step 1: Load the dataset
print("\nStep 1: Loading the UNSW-NB15 dataset...")
dataset = pd.read_csv(io.BytesIO(uploaded[filename]))

# Display basic information about the dataset
print(f"\nDataset shape: {dataset.shape}")
print(f"Number of rows: {dataset.shape[0]}")
print(f"Number of columns: {dataset.shape[1]}")

# Create a backup of the original dataset
dataset_original = dataset.copy()

# Step 2: Identify categorical columns
print("\nStep 2: Identifying categorical columns...")
categorical_columns = ['proto', 'service', 'state', 'attack_cat']

# Display information about categorical columns
print("\nCategorical columns in the dataset:")
categorical_info = pd.DataFrame({
    'Column': categorical_columns,
    'Data Type': [dataset[col].dtype for col in categorical_columns],
    'Unique Values': [dataset[col].nunique() for col in categorical_columns],
    'Sample Values': [', '.join(dataset[col].unique()[:5].astype(str)) + ('...' if dataset[col].nunique() > 5 else '') for col in categorical_columns]
})

display(categorical_info)

# For each categorical column, display the value distribution
print("\nDistribution of values in categorical columns:")
for col in categorical_columns:
    plt.figure(figsize=(12, 6))

    # For columns with many unique values, show only top values
    if dataset[col].nunique() > 15:
        value_counts = dataset[col].value_counts().nlargest(15)
        value_counts['Others'] = dataset[col].value_counts().nsmallest(dataset[col].nunique() - 15).sum()
    else:
        value_counts = dataset[col].value_counts()

    ax = sns.barplot(x=value_counts.index, y=value_counts.values)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')

    # Add count labels on top of bars
    for i, v in enumerate(value_counts.values):
        ax.text(i, v + 5, f"{v}", ha='center')

    plt.tight_layout()
    plt.show()

    # Display value distribution in a table
    value_percent = (dataset[col].value_counts() / len(dataset) * 100).round(2)
    value_distribution = pd.DataFrame({
        'Value': dataset[col].value_counts().index,
        'Count': dataset[col].value_counts().values,
        'Percentage (%)': value_percent.values
    })
    print(f"\nValue distribution for {col}:")
    display(value_distribution)

# Step 3: Choose encoding method based on the categorical column
print("\nStep 3: Applying different encoding methods based on the column characteristics...")

# Create a DataFrame to hold the encoded data
encoded_dataset = dataset.copy()

# 1. Label Encoding for binary or ordinal categories
print("\n1. Applying Label Encoding for binary or ordinal categories...")
label_encoders = {}

# Label encoding for attack_cat (since it directly corresponds to label)
label_encoders['attack_cat'] = LabelEncoder()
encoded_dataset['attack_cat_encoded'] = label_encoders['attack_cat'].fit_transform(encoded_dataset['attack_cat'])

# Display mapping for label encoding
attack_cat_mapping = pd.DataFrame({
    'Original': label_encoders['attack_cat'].classes_,
    'Encoded': range(len(label_encoders['attack_cat'].classes_))
})
print("\nLabel Encoding Mapping for attack_cat:")
display(attack_cat_mapping)

# 2. One-Hot Encoding for nominal categories with few unique values
print("\n2. Applying One-Hot Encoding for nominal categories with few unique values...")

# One-hot encoding for service and state (which have few unique values)
for col in ['service', 'state']:
    # Create dummies and drop first to avoid multicollinearity
    dummies = pd.get_dummies(encoded_dataset[col], prefix=col, drop_first=False)
    # Join the dummies to the original dataset
    encoded_dataset = pd.concat([encoded_dataset, dummies], axis=1)

    # Display the first few dummy columns
    print(f"\nFirst few one-hot encoded columns for {col}:")
    display(dummies.head())

# 3. Frequency Encoding for high-cardinality features
print("\n3. Applying Frequency Encoding for high-cardinality features...")
# Frequency encoding for proto (which has many unique values)
freq_map = encoded_dataset['proto'].value_counts(normalize=True)
encoded_dataset['proto_freq'] = encoded_dataset['proto'].map(freq_map)

# Display the frequency mapping for top protocols
print("\nFrequency Encoding for top protocols:")
top_protocols = freq_map.nlargest(10)
display(pd.DataFrame({'Protocol': top_protocols.index, 'Frequency': top_protocols.values}))

# Step 4: Decide which original columns to keep
print("\nStep 4: Creating the final encoded dataset...")
# List of original categorical columns to drop
columns_to_drop = categorical_columns.copy()

# Keep categorical columns if needed for analysis, reporting or visualization
# Uncomment the following line if you want to keep the original categorical columns
# columns_to_drop = []

# Create the final encoded dataset
if columns_to_drop:
    final_encoded_dataset = encoded_dataset.drop(columns=columns_to_drop)
    print(f"\nRemoved original categorical columns: {columns_to_drop}")
else:
    final_encoded_dataset = encoded_dataset
    print("\nKept all original categorical columns for reference")

# Display information about the final encoded dataset
print(f"\nFinal encoded dataset shape: {final_encoded_dataset.shape}")
print(f"Number of new columns added: {final_encoded_dataset.shape[1] - dataset.shape[1] + len(columns_to_drop)}")

# Display the first few rows of the encoded dataset
print("\nFirst few rows of the encoded dataset:")
display(final_encoded_dataset.head())

# Step 5: Save the encoded dataset
print("\nStep 5: Saving the encoded dataset...")
encoded_filename = 'UNSW_NB15_encoded.csv'
final_encoded_dataset.to_csv(encoded_filename, index=False)
print(f"Encoded dataset saved as '{encoded_filename}'")

# Allow user to download the encoded dataset
files.download(encoded_filename)

# Step 6: Generate encoding summary
print("\nStep 6: Generating encoding summary...")
encoding_summary = f"""
UNSW-NB15 Dataset Categorical Encoding Summary
==============================================

1. Original Dataset Information
   - Number of rows: {dataset_original.shape[0]}
   - Number of columns: {dataset_original.shape[1]}
   - Categorical columns: {categorical_columns}

2. Encoding Methods Applied
   - Label Encoding: Used for 'attack_cat' (direct mapping to target variable)
   - One-Hot Encoding: Used for 'service' and 'state' (nominal variables with few categories)
   - Frequency Encoding: Used for 'proto' (high-cardinality variable with many unique values)

3. Encoding Results
   - 'attack_cat': Encoded as numerical labels (0-{len(label_encoders['attack_cat'].classes_)-1})
   - 'service': Transformed into {encoded_dataset.filter(like='service_').shape[1]} binary columns
   - 'state': Transformed into {encoded_dataset.filter(like='state_').shape[1]} binary columns
   - 'proto': Transformed into frequency values (range: {encoded_dataset['proto_freq'].min():.4f}-{encoded_dataset['proto_freq'].max():.4f})

4. Final Encoded Dataset
   - Number of rows: {final_encoded_dataset.shape[0]} (unchanged)
   - Number of columns: {final_encoded_dataset.shape[1]}
   - New columns added: {final_encoded_dataset.shape[1] - dataset.shape[1] + len(columns_to_drop)}
   - Original categorical columns {"removed" if columns_to_drop else "retained"}

5. Next Steps
   - Normalize/standardize numerical features
   - Perform feature engineering
   - Address class imbalance if necessary
"""

print(encoding_summary)

# Save encoding summary to file
summary_filename = 'encoding_summary.txt'
with open(summary_filename, 'w') as f:
    f.write(encoding_summary)

# Allow user to download the summary
files.download(summary_filename)

print("\nCategorical encoding completed successfully!")
print("You can now proceed to the next preprocessing steps: normalization/standardization and feature engineering.")
```
