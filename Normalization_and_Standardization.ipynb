```python
# TPU-Optimized Standardization for UNSW-NB15 Dataset (Single Output)
# ------------------------------------------------------------------

# Install TensorFlow if not available
import sys
!pip install tensorflow

# Import necessary libraries
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
from google.colab import files
from IPython.display import display, HTML

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# Initialize TPU
try:
    tpu = tf.distribute.cluster_resolver.TPUClusterResolver()  # TPU detection
    print('Running on TPU ', tpu.cluster_spec().as_dict()['worker'])
    tf.config.experimental_connect_to_cluster(tpu)
    tf.tpu.experimental.initialize_tpu_system(tpu)
    strategy = tf.distribute.TPUStrategy(tpu)
    print("Number of accelerators: ", strategy.num_replicas_in_sync)
except ValueError:
    print('Not running on TPU, using default strategy')
    strategy = tf.distribute.get_strategy()

# Upload the encoded dataset
print("Please upload your encoded UNSW-NB15 dataset CSV file...")
uploaded = files.upload()

# Get the name of the uploaded file
filename = list(uploaded.keys())[0]
print(f"Processing file: {filename}")

# Step 1: Load the dataset
print("\nStep 1: Loading the encoded UNSW-NB15 dataset...")
dataset = pd.read_csv(io.BytesIO(uploaded[filename]))

# Display basic information about the dataset
print(f"\nDataset shape: {dataset.shape}")
print(f"Number of rows: {dataset.shape[0]}")
print(f"Number of columns: {dataset.shape[1]}")

# Step 2: Identify numerical features to standardize
print("\nStep 2: Identifying numerical features to standardize...")

# Get all numeric columns except 'id' and 'label'
numerical_columns = dataset.select_dtypes(include=['int64', 'float64']).columns.tolist()
columns_to_exclude = ['id', 'label', 'attack_cat_encoded']

# Remove excluded columns from the list of numerical columns
numerical_features = [col for col in numerical_columns if col not in columns_to_exclude]

print(f"\nIdentified {len(numerical_features)} numerical features to standardize:")
print(numerical_features)

# Step 3: Prepare data for TPU processing
print("\nStep 3: Preparing data for TPU processing...")

# Convert data to TensorFlow Dataset for TPU optimization
X = dataset[numerical_features].values.astype(np.float32)
X_tensor = tf.constant(X)

print(f"Tensor shape: {X_tensor.shape}")

# Step 4: TPU-optimized standardization (ONLY)
print("\nStep 4: Applying TPU-optimized standardization (ONLY)...")

with strategy.scope():
    @tf.function  # Use tf.function for TPU compilation
    def standardize_tpu(tensor):
        """Standardize features using TPU-optimized operations"""
        # Calculate mean and std along axis 0 (features axis)
        mean = tf.reduce_mean(tensor, axis=0, keepdims=True)
        std = tf.math.reduce_std(tensor, axis=0, keepdims=True)

        # Handle zero std values to avoid division by zero
        std = tf.where(tf.equal(std, 0), tf.ones_like(std), std)

        # Standardize
        standardized = (tensor - mean) / std

        return standardized, mean, std

    # Apply standardization
    standardized_tensor, mean_tensor, std_tensor = standardize_tpu(X_tensor)

    # Convert back to numpy for further processing
    standardized_array = standardized_tensor.numpy()
    mean_values = mean_tensor.numpy().flatten()
    std_values = std_tensor.numpy().flatten()

# Step 5: Create standardized dataset
print("\nStep 5: Creating standardized dataset...")

# Create a copy of the original dataset
standardized_dataset = dataset.copy()

# Replace the numerical features with standardized versions
standardized_dataset[numerical_features] = standardized_array

# Step 6: Visualize the results
print("\nStep 6: Visualizing the results...")

# Create a summary of the numerical features
numerical_summary = pd.DataFrame({
    'Feature': numerical_features,
    'Original Min': [dataset[col].min() for col in numerical_features],
    'Original Max': [dataset[col].max() for col in numerical_features],
    'Original Mean': [dataset[col].mean() for col in numerical_features],
    'Original Std': [dataset[col].std() for col in numerical_features],
    'Standardized Mean': [standardized_dataset[col].mean() for col in numerical_features],
    'Standardized Std': [standardized_dataset[col].std() for col in numerical_features]
})

print("\nSummary statistics for numerical features:")
display(numerical_summary)

# Visualize the distribution of key numerical features
key_features = ['dur', 'sbytes', 'dbytes', 'rate', 'sttl', 'dttl']
key_features = [f for f in key_features if f in numerical_features]

# Create comparison of distributions (original vs standardized)
plt.figure(figsize=(20, 10))
for i, feature in enumerate(key_features[:6]):
    plt.subplot(2, 3, i+1)
    sns.kdeplot(dataset[feature], color='blue', label='Original', alpha=0.7)
    sns.kdeplot(standardized_dataset[feature], color='red', label='Standardized', alpha=0.7)
    plt.title(f'{feature}')
    plt.legend()
plt.tight_layout()
plt.show()

# Step 7: Save the standardized dataset (SINGLE OUTPUT)
print("\nStep 7: Saving the standardized dataset...")

# Save single output file
output_filename = 'UNSW_NB15_preprocessed.csv'
standardized_dataset.to_csv(output_filename, index=False)

print(f"Standardized dataset saved as '{output_filename}'")

# Download the single output file
files.download(output_filename)

# Step 8: Generate preprocessing summary
print("\nStep 8: Generating preprocessing summary...")

preprocessing_summary = f"""
UNSW-NB15 Dataset TPU-Optimized Preprocessing Summary
===================================================

TPU Configuration:
- Accelerators available: {strategy.num_replicas_in_sync}
- TPU cluster detected: {'Yes' if 'tpu' in locals() else 'No'}

1. Original Dataset Information
   - Number of rows: {dataset.shape[0]}
   - Number of columns: {dataset.shape[1]}
   - Numerical features processed: {len(numerical_features)}

2. Preprocessing Steps Applied
   - Standardization: Applied using TPU acceleration
   - Method: Z-score standardization (mean=0, std=1)
   - All numerical features scaled to same unit
   - Categorical features preserved from encoding step

3. TPU Processing Performance
   - Strategy: {strategy.__class__.__name__}
   - Tensor shape processed: {X_tensor.shape}
   - Operations: Vectorized for optimal performance

4. Feature Statistics Summary (first 5 features)
   Original Range: Min={[round(float(x), 4) for x in numerical_summary['Original Min'][:5]]}, Max={[round(float(x), 4) for x in numerical_summary['Original Max'][:5]]}
   Standardized: Mean={[round(float(x), 10) for x in numerical_summary['Standardized Mean'][:5]]}, Std={[round(float(x), 10) for x in numerical_summary['Standardized Std'][:5]]}

5. Output
   - Single file produced: {output_filename}
   - Contains all features with standardized numerical values
   - Ready for ML model training

6. Next Steps
   - Use this preprocessed dataset for machine learning model training
   - Recommend for: SVM, Logistic Regression, Neural Networks
   - Consider feature engineering if additional improvements needed
   - Address class imbalance if necessary
"""

print(preprocessing_summary)

# Save preprocessing summary to file
summary_filename = 'preprocessing_summary.txt'
with open(summary_filename, 'w') as f:
    f.write(preprocessing_summary)

# Allow user to download the summary
files.download(summary_filename)

print("\nTPU-optimized preprocessing completed successfully!")
print("You now have a single preprocessed dataset ready for ML model training.")
```
